```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import numpy as np
import pandas as pd
```

```python
app = dash.Dash()
```

```python
df = pd.read_excel('D://DS//superstore//orders.xlsx')
df.head()
```

```python
app.layout = html.Div([
    
    dcc.Graph(
        id = 'scatter-chart',
        figure = {
            'data': [
                go.Scatter(
                    x = df.Sales,
                    y = df.Profit,
                    mode = 'markers'
                )
            ], 'layout': go.Layout(
                        title = 'scatterplot',
                        xaxis = {'title':'Sales'},
                        yaxis = {'title':'Profit'},
                        hovermode='closest'
            ) 
        }
    )
])
```

```python
if __name__ == '__main__':
    app.run_server(port=4051)
```
